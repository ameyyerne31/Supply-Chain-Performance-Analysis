```python
# Separate numerical and categorical columns
numerical_cols = SC.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = SC.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
#create count plot for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(10, 6))  # Adjust figure size
    sns.countplot(x=col, data=SC, palette='viridis')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
# Generate a list of colors from the 'coolwarm' palette
palette = sns.color_palette("viridis", n_colors=len(numerical_cols))

# Plotting histograms for each numerical column
for column, color in zip(numerical_cols, palette):
    plt.figure()  # Creates a new figure
    SC[column].hist(bins=10, color=color)  # Set color for each histogram
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    # Calculate mean, median, and mode
    mean_value = SC[column].mean()
    median_value = SC[column].median()
    mode_value = SC[column].mode().values[0]  # Handling multimodal data
    
    # Add vertical lines for mean, median, and mode
    plt.axvline(mean_value, color='r', linestyle='dashed', linewidth=1, label='Mean')
    plt.axvline(median_value, color='g', linestyle='dashed', linewidth=1, label='Median')
    plt.axvline(mode_value, color='b', linestyle='dashed', linewidth=1, label='Mode')
    
    # You can also add text annotations instead of lines
    # plt.text(mean_value, 100, f'Mean: {mean_value}', rotation=90)
    # plt.text(median_value, 100, f'Median: {median_value}', rotation=90)
    # plt.text(mode_value, 100, f'Mode: {mode_value}', rotation=90)
    
    plt.legend()
    plt.show()
```

```python
# Using describe on numerical columns
description = SC[numerical_cols].describe()

print(description)
```

```python
# Compute the correlation matrix
corr = SC[numerical_cols].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))  # Size of the figure
sns.heatmap(corr, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Extract the correlations between independent variables and the dependent variable
correlation_with_dependent = corr['Revenue generated']

# Sort the correlations in descending order
sorted_correlations = correlation_with_dependent.abs().sort_values(ascending=False)

# Display the attributes in descending order of correlation with the dependent variable
for attribute in sorted_correlations.index:
    correlation_percent = sorted_correlations[attribute] * 100
    print(f'{attribute} and dependent variable = {correlation_percent:.2f}%')
```

```python
# Define the pairs of numerical variables for joint plots
jointplot_pairs = [('Stock levels', 'Lead times'),
                   ('Order quantities', 'Production volumes'),
                   ('Shipping times', 'Shipping costs'),
                   ('Manufacturing lead time', 'Manufacturing costs'),
                   ('Number of products sold', 'Revenue generated'),
                   ('Availability', 'Defect rates')]

# Create joint plots for each pair of numerical variables
for pair in jointplot_pairs:
    sns.jointplot(data=SC, x=pair[0], y=pair[1], kind='scatter', marginal_ticks=True)
    plt.suptitle(f'{pair[0]} vs {pair[1]}', y=1.02)
    plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd


# Create a list of unique product types
product_types = SC['Product type'].unique()

# Create a figure
fig = go.Figure()

# Iterate through each product type
for product_type in product_types:
    # Filter data for the current product type
    filtered_data = SC[SC['Product type'] == product_type]
    
    # Add trace for availability
    fig.add_trace(go.Scatter(x=filtered_data.index, y=filtered_data['Availability'],
                             mode='lines+markers', name=f'Availability - {product_type}'))
    
    # Add trace for stock levels
    fig.add_trace(go.Scatter(x=filtered_data.index, y=filtered_data['Stock levels'],
                             mode='lines+markers', name=f'Stock Levels - {product_type}'))

# Update layout
fig.update_layout(title='Availability and Stock Levels Over Time by Product Type',
                  xaxis_title='Time',
                  yaxis_title='Percentage / Units',
                  legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01))

# Show interactive plot
fig.show()
```
